```python
import pandas as pd
import numpy as np
import datetime
```

```python
df = pd.read_csv('sales.csv',
                       sep=';', encoding='PT154', index_col='client_id')
```

```python
df.head()
```

```python
del df['level3']
```

```python
del df['order_number']
```

```python
df.head()
```

```python
df['item_quantity'] = df['item_quantity'].replace(
    to_replace=r',', value='.', regex=True).astype('float32')
df['item_sum'] = df['item_sum'].replace(
    to_replace=r',', value='.', regex=True).astype('float32')
```

```python
res_df = df.groupby(['client_id', 'level1']).agg(
    {'order_date': 'count', 'item_sum': 'sum'})\
.reset_index()\
.rename(columns={'order_date': 'order_freq', 'level1': 'items_group'})
res_df.head(20)
```
